```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
number_of_players= purchase_data_df["SN"].count()
number_of_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_unique_items = len(purchase_data_df["Item ID"].unique())
num_unique_items

revenue = purchase_data_df["Price"].sum()
revenue

average_price= purchase_data_df["Price"].mean()
average_price

purchases= purchase_data_df["Item ID"].count()
purchases

purchasing_analysis = pd.DataFrame([{"Unique Items": num_unique_items, "Total Revenue": revenue, 
                                     "Average Price of Item": average_price, "Total Purchases":purchases}])
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count_df= pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_count_df

percentage_of_gender = pd.DataFrame(purchase_data_df["Gender"].value_counts()/number_of_players*100)
percentage_of_gender

gender_count_df["Percentage of Gender"] = percentage_of_gender
gender_count_df["Percentage of Gender"] = gender_count_df["Percentage of Gender"].map("{:,.2f}%".format)
gender_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchasing_analysis_gender_df = purchase_data_df.groupby(["Gender"])
purchasing_analysis_gender_df

purchasing_analysis_gender_df["Purchase ID"].count().head(10)

purchasing_analysis_gender_df["Price"].mean().head(10)

total_purchase_by_gender = purchasing_analysis_gender_df["Price"].sum()
total_purchase_by_gender.head()

avg_purchase_price_gender = total_purchase_by_gender.map("${:,.2f}".format)
avg_purchase_price_gender.head()

summary_purchasing_analysis = pd.DataFrame(purchasing_analysis_gender_df["Purchase ID"].count())
summary_purchasing_analysis["Average Purchase by Gender"] = avg_purchase_price_gender 
summary_purchasing_analysis["Total Purchase by Gender"] = total_purchase_by_gender
summary_purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pd.cut(purchase_data_df["Age"], age_bins, labels = age_groups)
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], age_bins, labels = age_groups)
group_by_age = purchase_data_df.groupby(["SN", "Age Group"])

game_players_by_age = pd.DataFrame(group_by_age.size())
game_age_group_df = pd.DataFrame(game_players_by_age.groupby(["Age Group"]).count())
game_age_group_df.columns = ["Total Count"]
game_age_group_df["Percentage of Players"] = round(100 *game_age_group_df["Total Count"]/game_age_group_df["Total Count"].sum(),2)
game_age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
game_agegroup_df = pd.DataFrame(purchase_data_df.groupby(["Age Group"]).count())
game_agegroup_sum_df = pd.DataFrame(purchase_data_df.groupby(["Age Group"]).sum())
game_age_group_df["Purchase Count"] = game_agegroup_df["Item ID"]
game_age_group_df["Average Purchase Price"] = round(game_agegroup_sum_df["Price"]/game_age_group_df["Purchase Count"],2).map("${:,.2f}".format)
game_age_group_df["Total Purchase Value"] = game_agegroup_sum_df["Price"]
game_age_group_df["Total Purchase Value"] = game_age_group_df["Total Purchase Value"].map("${:,.2f}".format)
game_age_group_df = game_age_group_df.drop(labels = ["Total Count", "Percentage of Players"], axis = 1)
game_age_group_df 
```
